```python
y = df['charges']
```

```python
x = df.drop('charges',axis=1)
```

```python
x.shape, y.shape
```

```python
x_ohe=pd.get_dummies(x)
```

```python
x_ohe.head()
```

```python
x_ohe.isnull().sum()
```

```python
x_ohe.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test =train_test_split(x_ohe,y,
                                                random_state=18,
                                                test_size = 0.3)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
p = lr.predict(X_test)
```

```python
lr.coef_
```

```python
lasso = Lasso()
```

```python
lasso.fit(X_train,Y_train)
```

```python
lasso.predict(X_test)
```

```python
ridge = Ridge()
```

```python
ridge.fit(X_train,Y_train)
```

```python
ridge.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
mse_lr = mean_squared_error(Y_test,p)
mse_lasso = mean_squared_error(Y_test,lasso.predict(X_test))
mse_ridge = mean_squared_error(Y_test,ridge.predict(X_test))
```

```python
print("LR",mse_lr)
print("Lasso",mse_lasso)
print("Ridge",mse_ridge)
```
